## Limpando os dados

### Alterando valores da coluna Gênero

```python
generos = data['Genero'].copy()
generos[generos== 'Sports'] = 1
generos[generos== 'Platform'] = 2
generos[generos== 'Racing'] = 3
generos[generos== 'Role-Playing'] = 4
generos[generos== 'Puzzle'] = 5
generos[generos== 'Misc'] = 6
generos[generos== 'Shooter'] = 7
generos[generos== 'Simulation'] = 8
generos[generos== 'Action'] = 9
generos[generos== 'Fighting'] = 10
generos[generos== 'Adventure'] = 11
generos[generos== 'Strategy'] = 12
data['Genero'] = generos
```

### Alterando valores da coluna Platform

```python
plataforma = data['Plataforma'].copy()
plataforma[plataforma == 'Wii'] = 1
plataforma[plataforma == 'NES'] = 2
plataforma[plataforma == 'GB'] = 3
plataforma[plataforma == 'DS'] = 4
plataforma[plataforma == 'X360'] = 5
plataforma[plataforma == 'PS3'] = 6
plataforma[plataforma == 'PS2'] = 7
plataforma[plataforma == 'SNES'] = 8
plataforma[plataforma == 'GBA'] = 9
plataforma[plataforma == 'PS4'] = 10
plataforma[plataforma == '3DS'] = 11
plataforma[plataforma == 'N64'] = 12
plataforma[plataforma == 'PS'] = 13
plataforma[plataforma == 'XB'] = 14
plataforma[plataforma == 'PC'] = 15
plataforma[plataforma == '2600'] = 16
plataforma[plataforma == 'PSP'] = 17
plataforma[plataforma == 'XOne'] = 18
plataforma[plataforma == 'WiiU'] = 19
plataforma[plataforma == 'GC'] = 20
plataforma[plataforma == 'GEN'] = 21
plataforma[plataforma == 'DC'] = 22
plataforma[plataforma == 'PSV'] = 23
plataforma[plataforma == 'SAT'] = 24
plataforma[plataforma == 'SCD'] = 25
plataforma[plataforma == 'WS'] = 26
plataforma[plataforma == 'NG'] = 27
plataforma[plataforma == 'TG16'] = 28
plataforma[plataforma == '3DO'] = 29
plataforma[plataforma == 'GG'] = 30
plataforma[plataforma == 'PCFX'] = 31
data['Plataforma'] = plataforma
```

### Alterando valores da coluna Classificação

```python
#1      E
#2      M
#3      T
#4      E10+
#5      K-A
#6      AO
#7      EC
#8      RP

classificacao = data['Classificacao'].copy()
classificacao[classificacao == 'E'] = 1
classificacao[classificacao == 'M'] = 2
classificacao[classificacao == 'T'] = 3
classificacao[classificacao == 'E10+'] = 4
classificacao[classificacao == 'K-A'] = 5
classificacao[classificacao == 'AO'] = 6
classificacao[classificacao == 'EC'] = 7
classificacao[classificacao == 'RP'] = 8
data['Classificacao'] = classificacao
```

### Alterando valores da coluna Publicadora

```python
x = data['Publicadora'].unique()
for item in range(1,len(x)+1):
    b = data[data['Publicadora']==x[item-1]]
    b['Publicadora'] = item
    data[data['Publicadora']==x[item-1]] = b

# len(data['Publicadora'].unique())
```

### Alterando valores da coluna Desenvolvedora

```python
x = data['Desenvolvedora'].unique()
for item in range(1,len(x)+1):
    b = data[data['Desenvolvedora']==x[item-1]]
    b['Desenvolvedora'] = item
    data[data['Desenvolvedora']==x[item-1]] = b
```

```python
data.head()
data.info()
```

### Preenchendo valores vazios de colunas quantitativas

```python
data[data['Pontuacao dos usuarios'] == 'tbd'] = 0

data['Plataforma'] = data['Plataforma'].fillna(0)
data['Ano de lancamento'] = data['Ano de lancamento'].fillna(0)
data['Genero'] = data['Genero'].fillna(0)
data['Publicadora'] = data['Publicadora'].fillna(0)
data['Vendas NA'] = data['Vendas NA'].fillna(0)
data['Vendas EU'] = data['Vendas EU'].fillna(0)
data['Vendas JP'] = data['Vendas JP'].fillna(0)
data['Outras vendas'] = data['Outras vendas'].fillna(0)
data['Vendas globais'] = data['Vendas globais'].fillna(0)
data['Pontuacao da critica'] = data['Pontuacao da critica'].fillna(0)
data['Contagem de criticas'] = data['Contagem de criticas'].fillna(0)
data['Pontuacao dos usuarios'] = data['Pontuacao dos usuarios'].fillna(0)
data['Contagem de usuarios'] = data['Contagem de usuarios'].fillna(0)
data['Desenvolvedora'] = data['Desenvolvedora'].fillna(0)
data['Classificacao'] = data['Classificacao'].fillna(0)

### 

data['Pontuacao dos usuarios'] = data['Pontuacao dos usuarios'].astype('float64')
data = data.dropna()
```

```python
data.info()
```

### Criando uma coluna com as médias de pontuação

```python
data['Media de pontuacao'] = 0.0
def media(x, y):
    return (x+y)/2
data['Media de pontuacao'] = data.apply(lambda row: media(row['Pontuacao da critica'], row['Pontuacao dos usuarios']), axis=1)
data.head()
```
